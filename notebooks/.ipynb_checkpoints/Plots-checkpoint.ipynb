```python
import pandas as pd
import seaborn as sns
import matplotlib
```

```python
data = pd.read_csv("HDBSCAN_braycurtis_4(innerJoinMasterUNF_10_200(trimmed)).csv", sep="@")
#select columns
data = data.loc[:, data.columns.isin(["Prepositional Phrase Percentage","NN Percentage","nmod:tmod", "appos","Linked Noun Percentage","Verb Subject Percentage","Plural Verb Percentage","nmod:on","VBG Percentage", "cluster"])]
#select rows with cluster values
data = data.loc[data['cluster'].isin([17,57,36,2,33,86,91,71,82,30,63,28,47,1])]
# data
```

```python
#plot function
sns.set(style="whitegrid", palette="muted")
dataz = pd.melt(data, "cluster", var_name="attribute")
plot = sns.swarmplot(x="attribute", y="value", hue="cluster",data = dataz)
fig = plot.get_figure()
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(22, 10.5, forward=True)
fig.savefig('Cluster_Plot.png', dpi=100)
```

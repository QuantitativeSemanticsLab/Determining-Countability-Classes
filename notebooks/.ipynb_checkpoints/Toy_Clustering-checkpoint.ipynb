```python
import numpy as np
import pandas as pd
import hdbscan
import networkx as nx
import operator
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('/Users/aeshaanwahlang/Documents/Linguistics/RelevantDependanciesVector/iris.data.csv')
dataz = pd.DataFrame(iris)
x = iris.drop(columns=['Iris-setosa'])
```

```python
met = 'braycurtis'
min = 2
clusterer = hdbscan.HDBSCAN(min_cluster_size = min, metric = met, min_samples=4, prediction_data=True)
clusters = clusterer.fit(x)
predic_data = clusters.prediction_data_
```

```python
# cluster details
res = clusters.labels_
unique, counts = np.unique(res, return_counts = True)
uni = dict(zip(unique, counts))
print(uni)
# del uni[-1] #remove noise labels
max = 0
total = 0
for key in uni:
    val = uni[key]
    if(val > max):
        max = val
    total += val
sorted_uni = sorted(uni.items(), key=operator.itemgetter(1))
print("Clusters found: " + str(res.max()+1))
noise = np.count_nonzero(res == -1)
noise2 = round(noise/len(res), 3)*100
print("Noise: " + str(noise) + " (" + str(noise2) + "%)")
print("Avarage cluster size: " + str(round(total/len(uni), 3)) + "\nLargest cluster: " + str(max))
```

```python
dataz['cluster'] = res
nf = dataz.loc[dataz['cluster'] >= 0].sort_values(['cluster','Iris-setosa'])[['Iris-setosa','cluster']]
print(nf.to_string())
```

```python
digraph = clusterer.condensed_tree_.to_networkx()
adj = digraph.adjacency()
nx.draw_networkx(digraph)
plt.show()
```

```python
#Celex EML parsing
def parseEml(path):
    eml ={}
    with open(path, "r") as file:
        for line in file.readlines():
            split = line.split("\\")
            word = split[1]
            if word in eml: eml[word].append(split)
            else: eml[word] = [split]
        
        return eml
x = parseEml("/Users/aeshaanwahlang/Documents/QuantitativeSemanticsData/celex/eml-1.txt")
print(x["when"])
```

```python
with open("/Users/aeshaanwahlang/Documents/QuantitativeSemanticsData/docs/denumerators.json", "r") as f:
    dnum = json.load(f)
```

```python
def hasDigit(s):
    return bool(re.search(r"\d", s))
def hasLetter(s):
    return bool(re.search(r"[a-z]", s))

numbers = []
others = []

for k in dnum:
    if dnum[k] > 100:
        if hasDigit(k) and not hasLetter(k): numbers.append(k)
        else: others.append(k)
print(len(numbers))
```

```python
fuzzyNums = []
for i in range(len(numbers)):
    if bool(re.search(r"-", numbers[i])):
        fuzzyNums.append(numbers[i])
```

```python
print(len(numbers))
for fn in fuzzyNums:
    numbers.remove(fn)
print(len(numbers))
```

```python
x = []
for n in numbers:
    if bool(re.search(r"/", n)):
        x.append(n)
for val in x:
    numbers.remove(val)
print(len(numbers))
```

```python
def denumCheck(val):
    checks = ['.', '/', '-', '\\', '$']
    for c in checks:
        if c in val: return False
    return True

cleanedDnums = {}
for k in dnums_new:
    if dnums_new[k] >= 100 and k not in check and denumCheck(k):
        cleanedDnums[k] = dnums_new[k]
```
